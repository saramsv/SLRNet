```python
import pandas as pd
from PIL import Image
from matplotlib import pyplot as plt
import random
import cv2
import numpy as np
import json

random.seed(0)
num_classes = 8
colors = [(random.randint(0, 255), random.randint(
    0, 255), random.randint(0, 255)) for _ in range(num_classes)]
```

```python
def color_img(paths):
    for path in paths:
        img = cv2.imread(path) # this is the annotation
        try:
            for i in range(len(colors)):
                img[:,:,0][np.where(img[:,:,0] == i)] = colors[i][0]
                img[:,:,1][np.where(img[:,:,1] == i)] = colors[i][1]
                img[:,:,2][np.where(img[:,:,2] == i)] = colors[i][2]

            plt.imshow(img)
            plt.show()
        except:
            pass
```

```python
with open("data/seqsWLen10WOneAnn.odgt", 'r') as f:
    data = f.readlines()
```

```python
def display(start, end, data):
    for i in range(start,end):
        print(i)
        line_info = data[i].split(":")
        color_img(json.loads(line_info[1].strip().replace(", \"fpath_segm\"", '')))
        color_img(json.loads(line_info[2].strip().replace(", \"width\"", '')))
```
