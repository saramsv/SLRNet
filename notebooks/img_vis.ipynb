```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
def resize_img(img):
    width = int(120)
    height = int(80)
    dim = (width, height)

    # resize image
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    return resized
```

```python
# Load an color image in grayscale
img0 = cv2.imread('../0.jpg', 1)+150
img1 = cv2.imread('../1.jpg', 1)+150
img2 = cv2.imread('../2.jpg', 1)+150
img3 = cv2.imread('../3.jpg', 1)+150
img4 = cv2.imread('../4.jpg', 1)+150
img5 = cv2.imread('../5.jpg', 1)+150

img0 = resize_img(img0)
img1 = resize_img(img1)
img2 = resize_img(img2)
img3 = resize_img(img3)
img4 = resize_img(img4)
img5 = resize_img(img5)
```

```python
im0 = cv2.imread('../0cbr.png', 1)
im0 = np.concatenate((img0, im0), axis=0)

im1 = cv2.imread('../1cbr.png', 1)
im1 = np.concatenate((img1, im1), axis=0)

im2 = cv2.imread('../2cbr.png', 1)
im2 = np.concatenate((img2, im2), axis=0)

im3 = cv2.imread('../3cbr.png', 1)
im3 = np.concatenate((img3, im3), axis=0)

im4 = cv2.imread('../4cbr.png', 1)
im4 = np.concatenate((img4, im4), axis=0)

im5 = cv2.imread('../5cbr.png', 1)
im5 = np.concatenate((img5, im5), axis=0)

f, axarr = plt.subplots(2,3, figsize=(15, 10))

plt.title("cbr")

axarr[0,0].imshow(im0)
axarr[0,0].set_title("0cbr")

axarr[0,1].imshow(im1)
axarr[0,1].set_title("1cbr")

axarr[0,2].imshow(im2)
axarr[0,2].set_title("2cbr")

axarr[1,0].imshow(im3)
axarr[1,0].set_title("3cbr")

axarr[1,1].imshow(im4)
axarr[1,1].set_title("4cbr")

axarr[1,2].imshow(im5)
axarr[1,2].set_title("5cbr")

plt.tight_layout()
plt.show();
```
